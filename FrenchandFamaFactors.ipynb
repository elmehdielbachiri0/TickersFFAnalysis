```python
import yfinance as yf
import pandas as pd
import numpy as np

# Get the list of tickers for S&P 500
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
tickers = df['Symbol'].tolist()

# Exclude 'BRK.B' and 'BF.B' from the tickers list
exclude_symbols = ['BRK.B', 'BF.B']
tickers = [ticker for ticker in tickers if ticker not in exclude_symbols]

# Create an empty DataFrame to store the log returns
log_returns = pd.DataFrame()

# Loop over each stock in the S&P 500
for ticker in tickers:
    # Download historical stock prices for the current stock from Yahoo Finance
    stock_data = yf.download(ticker, start='2010-01-01', end='2023-05-11')
    
    # Calculate log returns as percentages
    stock_data['Log_Returns'] = np.log(stock_data['Adj Close'] / stock_data['Adj Close'].shift(1)) * 100
    
    # Append the log returns to the DataFrame
    log_returns[ticker] = stock_data['Log_Returns']

# Remove the first row with NaN values
log_returns = log_returns.dropna()

# Display the transformed log returns DataFrame
print(log_returns.head())
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Download the Fama-French factors data from the Fama/French website
ff_factors_url = 'https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_5_Factors_2x3_daily_CSV.zip'
ff_factors = pd.read_csv(ff_factors_url, skiprows=3, skipfooter=2, engine='python')

# Extract the necessary factors (RM, SMB, HML, RMW, CMA, RF)
ff_factors = ff_factors.rename(columns={'Unnamed: 0': 'Date'})
ff_factors['Date'] = pd.to_datetime(ff_factors['Date'], format='%Y%m%d')
ff_factors = ff_factors.set_index('Date')

```

```python

# Merge the log returns data with the Fama-French factors data
merged_data = pd.merge(log_returns, ff_factors, left_index=True, right_index=True)
merged_data.dropna()
# Display the merged data
print(merged_data.head())
# Exclude 'BRK.B' and 'BF.B' from the tickers list
exclude_symbols = ['BRK.B', 'BF.B']
tickers = [ticker for ticker in tickers if ticker not in exclude_symbols]
```

```python
import pandas as pd
import yfinance as yf
import numpy as np
import statsmodels.api as sm

results_matrix = np.zeros((501, 13))


# Loop over each stock in the S&P 500
for i, ticker in enumerate(tickers):
    # Get the excess returns for the current stock
    excess_returns = merged_data[ticker] - merged_data['RF_x']  # Subtract risk-free rate from the stock returns
    # Prepare the data for regression
    data = pd.DataFrame({'ticker': ticker,
        'excess_returns': excess_returns,
        'RM-RF': merged_data['Mkt-RF']-merged_data['RF_x'],
        'SMB': merged_data['SMB'],
        'HML': merged_data['HML'],
        'RMW': merged_data['RMW'],
        'CMA': merged_data['CMA_x'],
        'RF': merged_data['RF_x']
    })
    data = data.dropna()  # Remove any rows with missing values
    
    # Perform the regression
    X = data[['RM-RF', 'SMB', 'HML', 'RMW', 'CMA']]
    X = sm.add_constant(X)
    y = data['excess_returns']
    model = sm.OLS(y, X)
    results1 = model.fit()
    
    # Store the regression results in the matrix
    results_matrix[i, 1] = results1.params['const']  # Estimated constant coefficient
    results_matrix[i, 2] = results1.tvalues['const']  # T-statistic for the constant
    results_matrix[i, 0] = results1.rsquared  # R-squared
    
    for j, factor in enumerate(['RM-RF', 'SMB', 'HML', 'RMW', 'CMA']):
        results_matrix[i, j+3] = results1.params[factor]  # Estimated coefficient for each factor
        results_matrix[i, j+8] = results1.tvalues[factor]  # T-statistic for each factor


# Create a DataFrame from the results matrix
results_df = pd.DataFrame(results_matrix, columns=['Rsquared','Constant', 'T-Stat_Constant','RM-RF', 'SMB', 'HML', 'RMW', 'CMA', 'T-Stat_RM-RF', 'T-Stat_SMB', 'T-Stat_HML', 'T-Stat_RMW', 'T-Stat_CMA'])

# Set the ticker symbol as the index
results_df.index = tickers


# Display the results
print(results_df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot nonparametric density estimates for each t-statistic
sns.set(style='whitegrid')

# Loop over each t-statistic column
for column in results_df.columns[8:]:
    # Create a kernel density plot
    sns.kdeplot(results_df[column], fill=True, label=column)

# Add a legend and labels to the plot
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```

The T-Stat for tickers are close to follow a Guaussian Distribution, this shows that explanation of the fluctuation in log_returns for any ticker, using these variables is relevant, the coefficient are not inherently and very corolated to the ticker but they are rather corolated to the state of the market, with the upcoming analysis, I expect to have an even finer guaussian distribution when the tickers will be grouped by sector



```python
# Create a dictionary to store the sector data
sector_data = {}
# Loop over each stock ticker
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        # Get the sector information if available
        if 'sector' in stock.info:
            sector = stock.info['sector']
            sector_data[ticker] = sector
    except:
        print(f"Skipping {ticker}: Data not available")


# Create a DataFrame from the sector data
sector_df = pd.DataFrame.from_dict(sector_data, orient='index', columns=['Sector'])

```

```python

# Merge the sector classification data with the results DataFrame
results_with_sector = pd.merge(results_df, sector_df, left_index=True, right_index=True)
print(results_with_sector.head())
# Group the results by sector and calculate the mean factor coefficients
mean_coefficients_by_sector = results_with_sector.groupby('Sector').mean()[['RM-RF', 'SMB', 'HML', 'RMW', 'CMA']]

# Display the mean coefficients by sector
print(mean_coefficients_by_sector)
```

```python
# Plot nonparametric density estimates for each t-statistic
sns.set(style='whitegrid')


# Loop over each t-statistic column
for sec in mean_coefficients_by_sector.index:
    # Loop over each sector
    #dataframe per sector
    data2=results_with_sector.loc[results_with_sector['Sector'] == sec]

    for column in results_df.columns[8:-1]: 
        # Create a kernel density plot
        sns.kdeplot(data2[column], fill=True, label=column)

    # Add a legend and labels to the plot
    plt.legend()
    print(f'Distribution for the Sector {sec}')
    plt.xlabel('Value')
    plt.ylabel('Density')

    # Show the plot
    plt.show()
```

As seen above with finer granularity, we can see that the parameters estimates are close to following a guaussian, which we can explain by the fact that industries of a same sector are affected with the same external circumstances. We can clearly see that the distribution varies as well from sector to sector.
